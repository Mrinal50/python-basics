```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
days = [1,2,3,4,5,6,7]
max_t=[50,51,52,48,45,49,50]
min_t=[43,42,40,44,33,34,37]
avg_t=[45,48,45,42,43,45,46]
```

```python
plt.title('weather')
plt.xlabel('Days')
plt.ylabel('Temperature')

plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t,label="avg")

plt.legend()
```

```python
plt.title('weather')
plt.xlabel('Days')
plt.ylabel('Temperature')

plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")       #legion location change.and add shadow on legion for better looks
plt.plot(days,avg_t,label="avg")

plt.legend(loc="upper right",shadow=True)
```

```python
plt.title('weather')
plt.xlabel('Days')
plt.ylabel('Temperature')

plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")       #legion location change.and add shadow on legion for better looks.best fit of legion is to write best or type 0.
plt.plot(days,avg_t,label="avg")

plt.legend(loc="best",shadow=True)
```

```python
plt.title('weather')
plt.xlabel('Days')
plt.ylabel('Temperature')

plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")      
plt.plot(days,avg_t,label="avg")

plt.legend(loc="best",shadow=True,fontsize='small')
```

```python
#pie chart by using matplotlib.
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
exp_vals=[1600,600,100,410,250]
exp_labels=['home rent','travel rent','food','phone bill','sports']
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
#plt axis make it equal and plt show remove unwanted texts
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
#auto pct  %.
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%.0f%%')
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%.0f%%',shadow=True, explode=[0,0,0,0,0,])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%.0f%%',shadow=True, explode=[0,1,0,0,0,])
plt.show()
```

```python
#save it as png file
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%.0f%%',shadow=True, explode=[0,1,0,0,0,])
plt.show()
plt.savefig('piechart.png')
```

```python
#for good viewof png file,bbox inches is sufficient to get good aspect ratio image..
```
